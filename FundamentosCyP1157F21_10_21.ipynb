# Slicing (rebanado) de cadenas

Permite seleccionar subcadenas con el operador

[inicio:stop:incremento]

El stop no se incluye, es por eso que debemos agregar una posici贸n extra.

### Slicing y valores por defecto

Inicio: por defecto siempre ser谩 0

Stop: Tama帽o del arreglo (indexado positivo) 0 ( en un indexado negativo)

Incremento: es 1 (indexado positivo)

```python
carrera = "Ingenier铆a en computaci贸n"
print( carrera[0:10:1] )
print( carrera[14:25:1] )
# lo mismo pero con indexado negativo
print( carrera[-11:-1:1] )
print( carrera[:10:] )
print( carrera[14:25:] )
print( carrera[-11: :1] )

# SOLUCIN PREVIA
print( "-----------")
print( carrera[-11:-1:1] )
print( carrera[-11: :1] )

# un ejercico m谩s para ver incremento
print( carrera[0:10:2] )
print( carrera[-1:-12:-1])
print( carrera[25:13:-1])
```

# Funci贸n print

### Es empleada para desplegar informaci贸n en la salida est谩ndar. Normalmente un monitor.
